### In this dataset we are trying to predict the Happiness score of countries based on different factors...

```python
#import the basic libraries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the csv file through pandas
df_happy=pd.read_csv("happy_score.csv")
df_happy
```

```python
df_happy.info()
```

```python
df_happy.describe()
```

# Visualization

```python
#correlation between happiness score and happniess rank is higly negative so we need to drop the happiness rank
plt.figure(figsize=(10,8))
sns.heatmap(df_happy.corr(),annot=True)
```

```python
#checking the top 10 country which column has major impact in happiness score
for col in df_happy.select_dtypes(include=["float64"]):
    sns.barplot(y=df_happy["Country"][0:10],x=df_happy[col])
    plt.show()
```

```python
#drawing boxplot just to get the idea of outliners and it shows there are some outliners.so we will remove it through z score
df_happy.plot(kind="box",subplots=True,layout=(3,4),figsize=(10,10))
```

```python
#dropping the columns which has not relation with the target or which has no effect on target by removing them
df_happy.drop(columns=["Country","Region","Happiness Rank"],inplace=True)
```

```python
df_happy.head()
```

```python
#after removing some columns just checking the correlation which tells that all are necessary fields
plt.figure(figsize=(10,8))
sns.heatmap(df_happy.corr(),annot=True)
```

```python
#just removing the outliners if any through zscore and we can see the shape of dataset before and after zscore applied that outliners are removed
from scipy.stats import zscore
z_score=abs(zscore(df_happy))
print(df_happy.shape)
happy=df_happy[(z_score<3).all(axis=1)]
print(happy.shape)
```

```python
happy.head()
```

```python
#splitting the data into x(independent variable) and y(target variable)
x=happy.drop("Happiness Score",axis=1)
y=happy["Happiness Score"]
```

```python
x
```

```python
y.head()
```

## Using MinMax Scaling technique to scale the data of x(independent variables) so that they remain in same scale

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
#data after scaling
x
```

# Training the model

```python
#importing some other libaray for model training
from sklearn.model_selection import train_test_split
```

```python
#splitting the data into train and test where test is given 20% of data 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
```

# Using the Regression models as the target variable is continious

# LinearRegression model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
```

```python
pred=lr.predict(x_test)
pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
#checking the score and error 
print("Score",r2_score(y_test,pred))
print("mean_absolute_error",mean_absolute_error(y_test,pred))
print("mean_squared_error",mean_squared_error(y_test,pred))
```

# SVM model

```python
#using SVM technique with different kernel to know the which gives good score
from sklearn.svm import SVR
kernellist=['linear','poly','rbf']
for i in kernellist:
    sv=SVR(kernel=i)
    sv.fit(x_train,y_train)
    print(sv.score(x_train,y_train))
```

# Regularization :: L1 & L2 Regularizatiom
L1---Lasso Regression,
L2---Ridge Regression

```python
#using the regularization because score are getting near to 1 so just to avoid over fitting,
#if any using this to get best score by decreasing or removing  the coefficients 
from sklearn.linear_model import Lasso,Ridge
```

```python
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
```

```python
ls.coef_
```

```python
plt.figure(figsize=(18,5))
plt.bar(['Standard Error', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual'],ls.coef_)
plt.show()
```

```python
rd=Ridge(alpha=0.0001)
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
```

```python
plt.figure(figsize=(18,5))
plt.bar(['Standard Error', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual'],rd.coef_)
plt.show()
```

```python
#elasticnet uses both lasso and ridge so trying this to check score
from sklearn.linear_model import ElasticNet
enr=ElasticNet(alpha=0.0001)
enr.fit(x_train,y_train)
enrpred=enr.predict(x_test)
print(enr.score(x_train,y_train))
enr.coef_
```

# Ensemble Techniques

### Now using ensemble algorithm to reduce bias and errors and provide stability

```python
#Randomforest as bagging technique
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
rfr.score(x_train,y_train)
```

```python
#AdaBoost as boosting technique
from sklearn.ensemble import AdaBoostRegressor
model=AdaBoostRegressor()
model.fit(x_train,y_train)
model.score(x_train,y_train)
```

## Random forest is giving good score. linear and svm model also giving score more than random forest but just to avoid overfitting and stable model we go with random forest

### So we choose Randomforest as a model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
rfr.score(x_train,y_train)
```

### saving the best model

```python
#importing the joblib library to save the model
import joblib
joblib.dump(rfr,"rfr.happiness_score.pkl")
```
