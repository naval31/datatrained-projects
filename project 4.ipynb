#  Feature Engineering for Model building

```python
#checking no of unique values in a column
df_hr.nunique()
```

```python
# column having no. of unique values=1,we will drop those column 
df_hr=df_hr.drop(['StandardHours','Over18','EmployeeCount'],axis=1)
```

```python
#shape of dataset after dropping some columns
df_hr.shape
```

```python
# now using labelencoder converting categorical values to numeric for further study pf dataset 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in df_hr.select_dtypes(exclude='int64'):
    df_hr[col]=le.fit_transform(df_hr[col])
```

```python
#dataset after using labelencoder
df_hr.head()
```

```python
#split the data of dependent and independent variable
df_x=df_hr.drop(['Attrition'],axis=1)
y=df_hr['Attrition']
```

```python
#lets check the skewness of data
df_x.skew()
```

```python
#removing skewness using log transform
for col in df_x.columns:
    if df_x[col].skew()>0.55:
        df_x[col]=np.log1p(df_x[col])
```

#### Removing outliners

```python
#removing outliners using zscore
from scipy.stats import zscore
z_score=np.abs(zscore(df_x))
print(df_x.shape)
df_xnew=df_x[(z_score<3).all(axis=1)]
print(df_xnew.shape)
```

- All the outliners are removed from the dataset

```python
df_xnew
```

# Model selection/predictions

```python
#scaling the input variables
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

```python
# Again checking the skewness
x.skew()
```

```python
# breaking our input and target variable into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
# importing model for classification
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

#importing metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
```

```python
#using classification algorithm with the help of loop
Model=[LogisticRegression(),GaussianNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier()]
for M in Model:
    M.fit(x_train,y_train)
    predm=M.predict(x_test)
    print('\n')
    print("accuracy score of ",M,'is',accuracy_score(y_test,predm))
    print('\n')
    print(confusion_matrix(y_test,predm))
    print('\n')
    print(classification_report(y_test,predm,zero_division=1))
    print('\n')
    print("-----------------------------------------------------------------------------------------------------------")
```

#### For all the model above the LogisticRegression is giving maximum score of 0.88

######  .

```python
#using the cross validation
from sklearn.model_selection import cross_val_score
Model=[LogisticRegression(),GaussianNB(),SVC(),DecisionTreeClassifier(), KNeighborsClassifier()]
for i in Model:
    scores=cross_val_score(i,x,y,cv=3)
    print("scores of ",i)
    print("mean score",scores.mean())
    print("Standard deviation",scores.std())
    print('\n-------------------------------------------------------------------------------------------------------')
```

# roc_curve

```python
# roc curve and roc_auc_score for all algorithms 
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
Model=[LogisticRegression(),GaussianNB(),SVC(probability=True),DecisionTreeClassifier(), KNeighborsClassifier()]
for m in Model:    
    m.fit(x_train,y_train)
    y_pred_prob=m.predict_proba(x_test)[:,1]
    tpr,fpr,thresholds=roc_curve(y_test,y_pred_prob)
    plt.plot([0,1],[0,1],'k--')
    plt.plot(tpr,fpr,label=m)
    plt.xlabel('False positive rate')
    plt.ylabel('True positive rate')
    plt.title(m)
    plt.show()
    auc_score=roc_auc_score(y_test,m.predict(x_test))
    print("auc_score",auc_score)
```

# Using ensemble techniques to boost score

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print(accuracy_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

```python
from sklearn.ensemble import AdaBoostClassifier
ad=AdaBoostClassifier(base_estimator=LogisticRegression(),n_estimators=100,learning_rate=0.1)
ad.fit(x_train,y_train)
ad_pred=ad.predict(x_test)
print(accuracy_score(y_test,ad_pred))
print(confusion_matrix(y_test,ad_pred))
print(classification_report(y_test,ad_pred))
```

### AdaBoostClassifier with having base_estimator of LogisticRegression giving the maximum score.


#### Best model

```python
#AdaBoostClassifier(base_estimator=LogisticRegression())
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
ad=AdaBoostClassifier(base_estimator=LogisticRegression(),n_estimators=100,learning_rate=0.1)
ad.fit(x_train,y_train)
ad_pred=ad.predict(x_test)
print("accuracy_score:",accuracy_score(y_test,ad_pred))
print('\n')
print("confusion_matrix:\n",confusion_matrix(y_test,ad_pred))
print('\n')
print("classification_report:\n",classification_report(y_test,ad_pred))
y_pred_prob=ad.predict_proba(x_test)[:,1]
tpr,fpr,thresholds=roc_curve(y_test,y_pred_prob)
plt.plot([0,1],[0,1],'k--')
plt.plot(tpr,fpr,label='AdaBoostClassifier')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('AdaBoostClassifier')
plt.show()
auc_score=roc_auc_score(y_test,ad.predict(x_test))
print("auc_score",auc_score)
```

- so __AdaBoostClassifier__ is giving __score__ of nearly __0.90__ and also giving __precision__ of __0.90__
