```python
#using log method to reduce skewness...after using it still one column(Magnesium) is found to be large.
#Now apply another method boxcoz on that column
import numpy as np
for col in  df_wine.columns:
    if df_wine.skew().loc[col]>0.55:
        df_wine[col]=np.log1p(df_wine[col])
```

```python
#after using log method still one column(Magnesium) is found to be large.
#Now in next step apply another method boxcoz on that column
df_wine.skew()
```

```python
from scipy.stats import zscore
z_score=abs(zscore(df_wine))
print(df_wine.shape)
df_wine_final=df_wine[(z_score<3).all(axis=1)]
print(df_wine_final.shape)
```

```python
df_x=df_wine_final.drop(columns=["Class"])
y=df_wine_final[["Class"]].values
```

```python
df_x.shape
```

```python
y=y.reshape(-1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
x
```

```python
df_x.skew()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
def mscore(alg,x,y):
    max_score=0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=.20)
        alg.fit(x_train,y_train)
        pred=alg.predict(x_test)
        a_scr=accuracy_score(pred,y_test)
        if a_scr>max_score:
            max_score=a_scr
            final_r_state=r_state
    print()
    print("max accuracy score for corresponding is ",final_r_state,"is",max_score)
    return final_r_state
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001],'kernel': ['rbf', 'poly', 'sigmoid']}
svc=SVC()
gsvc=GridSearchCV(svc,param_grid,cv=10)
gsvc.fit(x,y)
gsvc.best_params_
```

```python
svc=SVC(C=10,gamma=0.001,kernel='rbf')
rstate=mscore(svc,x,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_para = {'criterion':['gini','entropy'],'max_depth': np.arange(3, 15)}
dtc=DecisionTreeClassifier()
dtg = GridSearchCV(dtc, tree_para, cv=5)
dtg.fit(x,y)
dtg.best_params_
```

```python
dtc=DecisionTreeClassifier(criterion='entropy',max_depth=3)
rstate=mscore(dtc,x,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighnors={"n_neighbors":range(1,30)}
knn=KNeighborsClassifier()
gknr=GridSearchCV(knn,neighnors,cv=10)
gknr.fit(x,y)
gknr.best_params_
```

```python
knn=KNeighborsClassifier(n_neighbors=23)
rstate=mscore(knn,x,y)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
knnscores=cross_val_score(knn,x,y,cv=5)
print(knnscores)
print("mean of the knn classifier: ",knnscores.mean())
print("STD of the knn classifier: ",knnscores.std())
```

```python
dtcscores=cross_val_score(dtc,x,y,cv=5)
print(knnscores)
print("mean of the dtc classifier: ",dtcscores.mean())
print("STD of the dtc classifier: ",dtcscores.std())
```

```python
svcscores=cross_val_score(svc,x,y,cv=5)
print(knnscores)
print("mean of the svc classifier: ",svcscores.mean())
print("STD of the svc classifier: ",svcscores.std())
```

```python
# boosting
from sklearn.ensemble import GradientBoostingClassifier
gbr=GradientBoostingClassifier()
param={'learning_rate':[0.001,0.01,0.1,1],'n_estimators':[10,50,100,150,200],"init":[svc,dtc,knn]}
ggb=GridSearchCV(gbr,param,cv=10)
ggb.fit(x,y)
ggb.best_params_
```

```python
gbr=GradientBoostingClassifier(init=knn,learning_rate=1,n_estimators=150)
print("mean score for gradient boosting classifier",cross_val_score(gbr,x,y,cv=5,scoring="balanced_accuracy").mean())
print("STD score for gradient boosting classifier",cross_val_score(gbr,x,y,cv=5,scoring="balanced_accuracy").std())    
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada_cls=AdaBoostClassifier()
param={'learning_rate':[0.001,0.01,0.1,1],'n_estimators':[10,50,100,150,200],"base_estimator":[svc,dtc,knn]}
ags=GridSearchCV(ada_cls,param,cv=10)
ags.fit(x,y)
ags.best_params_
```

```python
ada_cls=AdaBoostClassifier(base_estimator=dtc,learning_rate=1,n_estimators=150)
print("mean score for AdaBoostClassifier",cross_val_score(ada_cls,x,y,cv=5,scoring="balanced_accuracy").mean())
print("STD score for AdaBoostClassifier",cross_val_score(ada_cls,x,y,cv=5,scoring="balanced_accuracy").std())    
```
